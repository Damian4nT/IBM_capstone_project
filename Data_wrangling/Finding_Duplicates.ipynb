### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here

#2.1 Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.

pattern=df.groupby(['MainBranch', 'Employment', 'RemoteWork']).size().reset_index()
pattern.columns=['MainBranch', 'Employment', 'RemoteWork', 'pattern_frequency']
sorted_pattern=pattern.sort_values('pattern_frequency', ascending=False).reset_index()
sorted_pattern.drop('index', axis=1, inplace=True)
print('number of combinantions of this 3 columns:', sorted_pattern.shape[0])
sorted_pattern.head(10)
```

```python
#2.2.1 Clarify that these represent multiple respondents with identical answers rather than duplicate records.

top5pattern=sorted_pattern.loc[0:5,['pattern_frequency']].sum()
number_of_rows=df.shape[0]

print('share of top 5 combinations in this 3 cloumns in all answers:', (top5pattern/number_of_rows)*100 )

# these 5 combinations, which is not surprising because in the MainBranch column the most popular answer is "I am a developer by profession," which is expected in a survey for programmers rhather than something strange.
#Regarding RemoteWork, there are three dominant options: Remote, Hybrid, and In-person. Since only these three options exist, it does not introduce variability, and the data tends to repeat itself.
#So, for these three columns, it is completely normal to have so many duplicates.


#2.2.2 Analyse which other columns frequently share identical values within these response-pattern groups.

based_row_number=sorted_pattern.shape[0]

def check_patterns(column):
    pattern1=df.groupby(['MainBranch', 'Employment', 'RemoteWork', column]).size().reset_index()
    pattern1.columns=['MainBranch', 'Employment', 'RemoteWork', column, 'pattern_frequency']
    sorted_pattern1=pattern1.sort_values('pattern_frequency', ascending=False).reset_index()
    rows_in_table=sorted_pattern1.shape[0]
    size_info=f'\n number of combinantions of this 3 columns and {column} : {rows_in_table} \n'
    new_rows=f'number of new rows: {rows_in_table-based_row_number} \n'
    sorted_pattern1.drop('index', axis=1, inplace=True)
    

    return( print('\n\n\n\n','\t\t\t\t', column.upper(),'\n\n'), print(size_info), print(new_rows))

interesting_columns=['Age', 'CodingActivities', 'EdLevel', 'LearnCode', 'YearsCode', 'DevType', 'Country','LanguageHaveWorkedWith','PlatformHaveWorkedWith', 'AISent'] #Selected specific columns to avoid iterating over all 114 columns, many of which contain mostly empty values

for col in interesting_columns:
    check_patterns(col)
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
## Write your code here

#3.1 Create visualizations to show the distribution of shared response patterns across different categories.
duplicated_values=[]

for column in interesting_columns:
    duplicates= df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork', column])]
    duplicated_values.append(duplicates.shape[0])

    
plt.bar( x=interesting_columns, height=duplicated_values)
plt.xlabel('category')
plt.ylabel('count')
plt.title(' distribution of shared response patterns across different categories')
plt.xticks(rotation=90)
plt.show()
```

```python
#3.2 Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment
my3columns_dup=df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'])]

country_duplicates=my3columns_dup.groupby('Country').size().reset_index()
country_duplicates.columns=['Country','Size']
sorted_countries=country_duplicates.sort_values('Size', ascending=False).head(15)

sorted_countries.plot(kind='bar',x='Country', y='Size')
plt.xlabel('country')
plt.ylabel('number of respondents')
plt.title('distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork grouped by Country', fontsize=15)
plt.show()

employment_duplicates=my3columns_dup.groupby('Employment').size().reset_index()
employment_duplicates.columns=['Employment','Size']
sorted_employment=employment_duplicates.sort_values('Size').tail(10)
exp=[0.2,0.15,0.1,0.05,0,0,0,0,0,0]

sorted_employment.plot(kind='pie',y='Size', figsize=(20,7), autopct='%1.1f%%', startangle=90, labels=None, pctdistance=1.1, explode=exp)
plt.title('distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork grouped by Employment', y=1.15, fontsize=15)
plt.legend(labels=sorted_employment['Employment'], loc='upper right')
plt.axis('equal')
plt.show()

```
