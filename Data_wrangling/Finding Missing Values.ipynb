<h2>1. Load the Dataset</h2>

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here

df.describe(include='all')
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here

#df.info(verbose=True, show_counts=True)

null_values=df.isnull().sum()

print(null_values[null_values>0])
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(8,25))
sns.heatmap(pd.DataFrame(null_values), cmap='YlOrRd', annot=True, annot_kws={'fontsize':11, 'fontweight': 'bold'})
plt.title('graphical null value presentation', y=1.02, fontsize=20)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here

RemoteWork_NaN=df['RemoteWork'].isnull().sum()
print('number of missing values for RemoteWork:', RemoteWork_NaN)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here

grouped_RemoteWork=df.groupby('RemoteWork').size().reset_index(name='size')

print(grouped_RemoteWork, '\n\n')

most_frequent_RemoteWork=df['RemoteWork'].mode()[0]

print('most_frequent_RemoteWork:\t',most_frequent_RemoteWork)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here

df['RemoteWork']=df['RemoteWork'].fillna(most_frequent_RemoteWork)

RemoteWork_NaN_final=df['RemoteWork'].isnull().sum()

print('RemoteWork number of NaN after imputation :', RemoteWork_NaN_final)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here

plt.hist(df['RemoteWork'],edgecolor='black', linewidth=1)
plt.xlabel('RemoteWork')
plt.ylabel('number of respondents')
plt.title('distribution of RemoteWork')
plt.show()
```
